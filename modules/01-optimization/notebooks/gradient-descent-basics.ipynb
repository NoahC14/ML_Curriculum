```python
from __future__ import annotations

import numpy as np

np.set_printoptions(precision=4, suppress=True)
```

## Step 1 - Objective and gradient

We minimize a one-dimensional quadratic. This is small enough to inspect directly, but still captures the standard gradient descent update pattern used in larger models.


```python
def objective(theta: float) -> float:
    return (theta - 3.0) ** 2 + 2.0


def gradient(theta: float) -> float:
    return 2.0 * (theta - 3.0)


theta = -4.0
learning_rate = 0.2
history = []

for step in range(12):
    history.append((step, theta, objective(theta), gradient(theta)))
    theta = theta - learning_rate * gradient(theta)

history[:4]
```

## Step 2 - Convergence summary

A well-chosen learning rate should decrease the objective steadily for this convex problem.


```python
objective_values = np.array([item[2] for item in history])
monotone_decrease = bool(np.all(np.diff(objective_values) <= 1e-12))

{
    "final_theta": theta,
    "final_objective": objective(theta),
    "monotone_decrease": monotone_decrease,
}
```

## Exercise

Compare the final iterate with the analytic minimizer and compute the absolute error.


```python
analytic_minimizer = 3.0
absolute_error = abs(theta - analytic_minimizer)

{"analytic_minimizer": analytic_minimizer, "absolute_error": absolute_error}
```
