## Exercise 1. Span and linear dependence

Solve `a1 v1 + a2 v2 + a3 v3 = 0`. A nontrivial solution shows dependence, and here `v3 = v1 + v2`.


```python
v1 = sp.Matrix([1, 0, 1])
v2 = sp.Matrix([0, 1, 1])
v3 = sp.Matrix([1, 1, 2])
a1, a2, a3 = sp.symbols("a1 a2 a3")
equations = list(a1 * v1 + a2 * v2 + a3 * v3)
sol = sp.solve(equations, [a1, a2], dict=True)
{"family": sol, "relation_check": sp.simplify(v3 - v1 - v2)}
```

## Exercise 2. Orthogonal projection

For a unit vector `u`, the projector onto its span is `P = u u^T`.


```python
u = sp.Matrix([sp.sqrt(sp.Rational(1, 2)), sp.sqrt(sp.Rational(1, 2))])
P = sp.simplify(u * u.T)
x = sp.Matrix([3, 1])
{"P": P, "Px": sp.simplify(P * x)}
```

## Exercise 3. Positive semidefinite covariance

For any vector `v`, `v^T Sigma v = E[((X - mu)^T v)^2] >= 0`. The code below verifies the square form symbolically.


```python
v1s, v2s, x1, x2, mu1, mu2 = sp.symbols("v1s v2s x1 x2 mu1 mu2", real=True)
v = sp.Matrix([v1s, v2s])
xm = sp.Matrix([x1 - mu1, x2 - mu2])
outer = xm * xm.T
quadratic = sp.expand((v.T * outer * v)[0])
square = sp.expand((xm.dot(v)) ** 2)
sp.simplify(quadratic - square)
```

## Exercise 4. Symmetric eigenvector orthogonality

If `A` is symmetric and `Au = lambda_u u`, `Av = lambda_v v`, then
`u^T A v = lambda_v u^T v = (Au)^T v = lambda_u u^T v`.
Distinct eigenvalues force `u^T v = 0`.


```python
A = sp.Matrix([[2, 1], [1, 2]])
eigs = A.eigenvects()
orthogonality = sp.simplify((sp.Matrix([1, 1]).dot(sp.Matrix([1, -1]))))
{"eigen_data": eigs, "dot_product_of_distinct_eigenvectors": orthogonality}
```

## Exercise 5. Spectral decomposition

## Exercise 6. First principal component from constrained optimization

The Lagrangian is `L(v, lambda) = v^T S v - lambda (v^T v - 1)`, and stationarity gives `Sv = lambda v`.


```python
s11, s12, s22, v1c, v2c, lam = sp.symbols("s11 s12 s22 v1c v2c lam", real=True)
S = sp.Matrix([[s11, s12], [s12, s22]])
v = sp.Matrix([v1c, v2c])
lagrangian_grad = [sp.diff((v.T * S * v)[0] - lam * ((v.T * v)[0] - 1), var) for var in (v1c, v2c)]
lagrangian_grad
```

## Exercise 9. Whitening transform


```python
Q = sp.Matrix([[1, 0], [0, 1]])
Lambda = sp.diag(4, 9)
Sigma = Q * Lambda * Q.T
W = Lambda ** sp.Rational(-1, 2) * Q.T
{"W": W, "whitened_covariance": sp.simplify(W * Sigma * W.T)}
```

## Exercise 10. Conditioning and regression

Near-collinearity means at least one singular value of `X` is close to zero, so `X^T X` is ill-conditioned and the normal-equation solve amplifies perturbations.


```python
X = sp.Matrix([[1, 1], [1, 1 + sp.Rational(1, 1000)], [1, 1 - sp.Rational(1, 1000)]])
XtX = sp.simplify(X.T * X)
{"X^T X": XtX, "determinant": sp.factor(XtX.det())}
```

## Exercise 11. Why SVD is more general than eigendecomposition

Eigenvectors require a square map from a vector space to itself. SVD handles rectangular linear maps by separating input directions (`V`), output directions (`U`), and axis-wise scaling (`Sigma`).
