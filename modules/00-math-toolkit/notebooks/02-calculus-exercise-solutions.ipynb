## Exercise 1. Partial derivatives and gradient


```python
x1, x2, x3 = sp.symbols("x1 x2 x3", real=True)
f = x1**2 * x2 + sp.exp(x3) - x1 * x3
{"partials": [sp.diff(f, var) for var in (x1, x2, x3)], "gradient": sp.Matrix([sp.diff(f, var) for var in (x1, x2, x3)])}
```

## Exercise 2. Directional derivative


```python
x1, x2 = sp.symbols("x1 x2", real=True)
f = x1**2 + 2 * x1 * x2 + 3 * x2**2
grad = sp.Matrix([sp.diff(f, x1), sp.diff(f, x2)])
point = {x1: 1, x2: -1}
u = sp.Matrix([2, 1])
{"grad_at_point": grad.subs(point), "directional_derivative": (grad.subs(point).dot(u))}
```

## Exercise 3. Jacobian of a vector-valued map


```python
x1, x2, x3 = sp.symbols("x1 x2 x3", real=True)
F = sp.Matrix([x1 * x2 + x3, x1**2 - sp.sin(x2 * x3)])
F.jacobian([x1, x2, x3])
```

## Exercise 4. Hessian and critical point classification


```python
x1, x2 = sp.symbols("x1 x2", real=True)
f = x1**2 + x1 * x2 + x2**2 - 4 * x1
grad = [sp.diff(f, var) for var in (x1, x2)]
critical = sp.solve(grad, [x1, x2], dict=True)[0]
hessian = sp.hessian(f, (x1, x2))
{"critical_point": critical, "hessian": hessian, "eigenvalues": hessian.eigenvals()}
```

## Exercise 5. Second-order Taylor approximation


```python
x1, x2 = sp.symbols("x1 x2", real=True)
f = sp.log(1 + x1 + x2)
series = sp.series(f.expand(), x1, 0, 3).removeO()
taylor2 = 1 * 0
for total_degree in range(0, 3):
    for i in range(total_degree + 1):
        j = total_degree - i
        deriv = sp.diff(f, x1, i, x2, j).subs({x1: 0, x2: 0})
        taylor2 += deriv * x1**i * x2**j / (sp.factorial(i) * sp.factorial(j))
sp.expand(taylor2)
```

## Exercise 6. Vector chain rule in coordinates


```python
x1, x2 = sp.symbols("x1 x2", real=True)
g = sp.Matrix([x1 + x2**2, x1 * x2])
y1, y2 = sp.symbols("y1 y2", real=True)
f = y1**2 + 3 * y2
h_direct = sp.expand((x1 + x2**2) ** 2 + 3 * x1 * x2)
grad_direct = sp.Matrix([sp.diff(h_direct, x1), sp.diff(h_direct, x2)])
Jg = g.jacobian([x1, x2])
grad_f = sp.Matrix([sp.diff(f, y1), sp.diff(f, y2)]).subs({y1: g[0], y2: g[1]})
grad_chain = sp.simplify(Jg.T * grad_f)
{"direct": grad_direct, "chain_rule": grad_chain, "difference": sp.simplify(grad_direct - grad_chain)}
```

## Exercise 7. Affine map plus squared loss


```python
m, d = 2, 3
W = sp.MatrixSymbol("W", m, d)
x = sp.MatrixSymbol("x", d, 1)
b = sp.MatrixSymbol("b", m, 1)
y = sp.MatrixSymbol("y", m, 1)
"For z = Wx + b and L = 1/2 ||z - y||^2, grad_z L = z - y, grad_b L = z - y, grad_W L = (z - y) x^T."
```

## Exercise 8. Affine map plus elementwise nonlinearity


```python
z1, z2, c1, c2, x1s, x2s = sp.symbols("z1 z2 c1 c2 x1s x2s", real=True)
a = sp.Matrix([sp.tanh(z1), sp.tanh(z2)])
c = sp.Matrix([c1, c2])
L = (c.T * a)[0]
grad_z = sp.Matrix([sp.diff(L, z1), sp.diff(L, z2)])
{"grad_z": grad_z, "pattern": "grad_W L = grad_z x^T"}
```

## Exercise 9. Hessian of least squares


```python
X = sp.MatrixSymbol("X", 4, 2)
w = sp.MatrixSymbol("w", 2, 1)
y = sp.MatrixSymbol("y", 4, 1)
"For f(w) = 1/2 ||Xw - y||^2, grad f = X^T (Xw - y) and Hessian = X^T X, which is PSD."
```

## Exercise 10. Newton step for a quadratic objective

## Exercise 11. Sphere constraint with Lagrange multipliers


```python
a1, a2, lam = sp.symbols("a1 a2 lam", real=True)
x1, x2 = sp.symbols("x1 x2", real=True)
L = a1 * x1 + a2 * x2 - lam * (x1**2 + x2**2 - 1)
[sp.diff(L, var) for var in (x1, x2, lam)]
```

## Exercise 12. PCA-style constrained optimization


```python
s11, s12, s22, v1, v2, lam = sp.symbols("s11 s12 s22 v1 v2 lam", real=True)
S = sp.Matrix([[s11, s12], [s12, s22]])
v = sp.Matrix([v1, v2])
stationarity = sp.Matrix([sp.diff((v.T * S * v)[0] - lam * ((v.T * v)[0] - 1), var) for var in (v1, v2)])
stationarity
```
