## Exercise 1. Verify the category axioms in Set


```python
f = lambda a: a + 1
g = lambda b: 2 * b
h = lambda c: c - 3
a = 5
{"left_assoc": h(g(f(a))), "right_assoc": (lambda x: h(g(x)))(f(a)), "left_identity": f(a), "right_identity": f(a)}
```

Associativity follows because both composites send `a` to `h(g(f(a)))`, and identity functions leave every element unchanged.


## Exercise 2. Linear maps as morphisms in Vect_R


```python
A = sp.Matrix([[1, 2], [0, 1], [1, -1]])
B = sp.Matrix([[2, 0, 1], [1, 1, 0]])
{"BA": B * A, "linearity_check": (B * A) * sp.Matrix([sp.Symbol("x1"), sp.Symbol("x2")])}
```

Composition of linear maps is linear, and its matrix is the product `BA`. This is the same algebra that stacked linear layers use.


## Exercise 3. A non-example from groups


For groups `(R, +)` to `(R, +)`, the function `f(x) = x^2` is not a homomorphism because `f(x + y) = x^2 + 2xy + y^2`, which is not generally equal to `f(x) + f(y)`.


## Exercise 4. Read a commutative diagram

## Exercise 5. Product universal property in Set


```python
f = lambda x: x + 1
g = lambda x: x * x
pair = lambda x: (f(x), g(x))
x = 3
{"pair(x)": pair(x), "pi_A(pair(x))": pair(x)[0], "pi_B(pair(x))": pair(x)[1]}
```

## Exercise 6. Coproduct case analysis


A disjoint union `A sqcup B` models a dataset where each example is tagged as either image or tabular. To define a map into labels `Y`, we need one branch map `A -> Y` and one branch map `B -> Y`.


## Exercise 7. Functor sanity check


The forgetful functor sends identity linear maps to identity functions and sends a composite linear map to the composite of its underlying functions, so it preserves identities and composition.


## Exercise 8. Naturality square


For `f : A -> B`, naturality is `G(f) ∘ eta_A = eta_B ∘ F(f)`. The components `eta_A` therefore fit together coherently across morphisms rather than being arbitrary unrelated maps.


## Exercise 9. Feature pipeline as composition


The full predictor is `c ∘ phi ∘ n`. Associativity means we may group it as `(c ∘ phi) ∘ n` or `c ∘ (phi ∘ n)` without changing the map, though we may not reorder the stages.


## Exercise 10. Label-preserving augmentation


Label preservation is the commuting condition `y ∘ a = y`. This matters because augmentation should transform inputs without silently changing the supervised target.


## Exercise 11. Encoder-decoder factorization


Introducing the latent object `Z` makes the bottleneck explicit. That isolates compression, representation structure, and interpretability in a way a single opaque map `X -> X` does not.


## Exercise 12. Transfer learning diagram


A transfer pipeline has composite `c_tgt ∘ t ∘ f_src`. The adaptation map `t` is needed when the source representation is useful but not already aligned with the target task or target feature geometry.


## Exercise 13. Monoidal intuition


A product-like or tensor-like construction captures parallel processing by sending two inputs through separate branches first and combining them only afterward. That distinguishes parallel composition from ordinary sequential composition.
