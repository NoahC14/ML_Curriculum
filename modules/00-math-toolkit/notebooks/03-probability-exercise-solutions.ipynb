## Exercise 1. PMF normalization


```python
c = sp.Rational(1, 6)
pmf = {0: c, 1: 2 * c, 2: 3 * c}
EX = sum(x * p for x, p in pmf.items())
EX2 = sum((x**2) * p for x, p in pmf.items())
{"c": c, "E[X]": sp.simplify(EX), "Var[X]": sp.simplify(EX2 - EX**2)}
```

## Exercise 3. Covariance is positive semidefinite


```python
v1, v2, x1, x2, mu1, mu2 = sp.symbols("v1 v2 x1 x2 mu1 mu2", real=True)
v = sp.Matrix([v1, v2])
centered = sp.Matrix([x1 - mu1, x2 - mu2])
sp.expand((v.T * centered * centered.T * v)[0] - (centered.dot(v)) ** 2)
```

## Exercise 4. Correlation and feature redundancy


Correlation near `0.99` means the design matrix has nearly collinear columns. That makes the smallest singular value small, destabilizes linear regression, and PCA can help by rotating into orthogonal coordinates and optionally discarding the weak redundant direction.


## Exercise 5. Joint, marginal, and conditional probabilities


```python
joint = {
    (1, 1): sp.Rational(3, 10),
    (1, 0): sp.Rational(1, 10),
    (0, 1): sp.Rational(1, 5),
    (0, 0): sp.Rational(2, 5),
}
P_Y1 = joint[(1, 1)] + joint[(0, 1)]
P_X1 = joint[(1, 1)] + joint[(1, 0)]
P_Y1_given_X1 = sp.simplify(joint[(1, 1)] / P_X1)
independent = sp.simplify(joint[(1, 1)] - P_Y1 * P_X1) == 0
{"P(Y=1)": P_Y1, "P(X=1)": P_X1, "P(Y=1|X=1)": P_Y1_given_X1, "independent": independent}
```

## Exercise 6. Bayes theorem for spam filtering


```python
posterior = sp.Rational(6, 25) / (sp.Rational(6, 25) + sp.Rational(1, 25))
{"posterior": sp.simplify(posterior), "decimal": float(posterior)}
```

A message containing `free` is more likely to be spam than a random message, but the posterior is still below one because non-spam messages can also contain the word.


## Exercise 7. Bernoulli MLE


```python
n, s, theta = sp.symbols("n s theta", positive=True)
loglik = s * sp.log(theta) + (n - s) * sp.log(1 - theta)
crit = sp.solve(sp.diff(loglik, theta), theta)[0]
second = sp.simplify(sp.diff(loglik, theta, 2).subs(theta, crit))
{"theta_hat": sp.simplify(crit), "second_derivative_at_hat": second}
```

## Exercise 8. Gaussian mean MLE with known variance


```python
n, sigma2, mu = sp.symbols("n sigma2 mu", positive=True)
x1, x2, x3 = sp.symbols("x1 x2 x3", real=True)
xs = [x1, x2, x3]
score = sp.diff(-sum((x - mu) ** 2 for x in xs) / (2 * sigma2), mu)
mu_hat = sp.solve(score, mu)[0]
mu_hat
```

## Exercise 9. Gaussian variance MLE


```python
n, sigma2 = sp.symbols("n sigma2", positive=True)
S = sp.symbols("S", nonnegative=True)
loglik = -sp.Rational(1, 2) * n * sp.log(sigma2) - S / (2 * sigma2)
sp.solve(sp.diff(loglik, sigma2), sigma2)[0]
```

The MLE uses denominator `n`; the unbiased estimator replaces it with `n - 1` to correct finite-sample bias after estimating the mean.


## Exercise 10. MAP as regularized estimation


```python
w, tau2 = sp.symbols("w tau2", positive=True)
residual = sp.symbols("residual", real=True)
objective = residual**2 + w**2 / tau2
objective
```

A zero-mean Gaussian prior adds an `l2` penalty, so MAP shrinks the parameter toward zero.

## Exercise 12. Dirichlet smoothing


```python
prior = sp.Matrix([1, 1, 1])
counts = sp.Matrix([15, 3, 2])
prior + counts
```

The posterior parameters are `(16, 4, 3)`, which is the usual add-one smoothing rule.


## Exercise 13. Poisson modeling decision


A Poisson model is appropriate for counts such as the number of clicks or arrivals in a fixed window. The count structure matters because the target is nonnegative and integer-valued, but the Poisson mean-variance tie can be too restrictive when data are overdispersed.


## Exercise 14. Generative versus discriminative framing


Discriminative modeling learns `p(y | x)` directly. Generative modeling learns `p(x | y)` with `p(y)` and uses Bayes' rule, which can exploit stronger distributional structure but requires modeling more of the data-generating process.
