```python
from __future__ import annotations

import numpy as np

np.set_printoptions(precision=3, suppress=True)
SEED = 7
rng = np.random.default_rng(SEED)
SEED
```

## Step 1 - Matrix-vector multiplication

The matrix below acts like a tiny linear feature map. In linear models, this is the core operation behind predictions and basis changes.


```python
A = np.array([[2.0, 1.0], [1.0, 3.0]])
x = np.array([1.5, -0.5])
Ax = A @ x

A, x, Ax
```

## Step 2 - Eigenvectors and invariant directions

Symmetric matrices appear throughout ML, especially in covariance analysis and quadratic objectives. Their eigenvectors provide orthogonal directions that simplify reasoning.


```python
eigenvalues, eigenvectors = np.linalg.eigh(A)
reconstruction_error = np.linalg.norm(A @ eigenvectors[:, 0] - eigenvalues[0] * eigenvectors[:, 0])

{"eigenvalues": eigenvalues, "reconstruction_error": reconstruction_error}
```

## Exercise

Standardization is one of the first matrix-shaped data transformations learners encounter. Compute column means and standard deviations for a tiny feature matrix.


```python
features = rng.normal(loc=[3.0, -1.0], scale=[2.0, 0.5], size=(5, 2))
means = features.mean(axis=0)
stds = features.std(axis=0)
standardized = (features - means) / stds

{
    "means": means,
    "stds": stds,
    "standardized_column_means": standardized.mean(axis=0),
}
```
