## Exercise 1. Bernoulli entropy


```python
p = sp.symbols("p", positive=True)
H = -p * sp.log(p) - (1 - p) * sp.log(1 - p)
{"H(p)": H, "dH/dp": sp.simplify(sp.diff(H, p)), "critical_point": sp.solve(sp.diff(H, p), p)}
```

The second derivative is negative on `(0,1)`, so `p = 1/2` gives the maximum entropy.


## Exercise 2. Cross-entropy decomposition


```python
p1, p2, q1, q2 = sp.symbols("p1 p2 q1 q2", positive=True)
H_P = -(p1 * sp.log(p1) + p2 * sp.log(p2))
KL = p1 * sp.log(p1 / q1) + p2 * sp.log(p2 / q2)
H_PQ = -(p1 * sp.log(q1) + p2 * sp.log(q2))
sp.expand(H_P + KL - H_PQ)
```

Minimizing cross-entropy in `Q` is the same as minimizing KL from `P` to `Q` because `H(P)` does not depend on `Q`.


## Exercise 3. Empirical cross-entropy


For one-hot labels only the correct class survives in each inner sum, so the average negative log-likelihood is exactly the empirical cross-entropy loss.


## Exercise 4. Nonnegativity of KL divergence


By Gibbs' inequality or Jensen's inequality for the concave logarithm, `D_KL(P || Q) >= 0`, with equality only when the two distributions agree on the support of `P`.


## Exercise 5. Bernoulli KL divergence


```python
p_val = sp.Rational(4, 5)
q1 = sp.Rational(1, 2)
q2 = sp.Rational(4, 5)
kl_expr = p * sp.log(p / sp.Symbol("q")) + (1 - p) * sp.log((1 - p) / (1 - sp.Symbol("q")))
kl_08_05 = sp.N(p_val * sp.log(p_val / q1) + (1 - p_val) * sp.log((1 - p_val) / (1 - q1)))
kl_08_08 = sp.N(p_val * sp.log(p_val / q2) + (1 - p_val) * sp.log((1 - p_val) / (1 - q2)))
{"formula": kl_expr, "KL(0.8||0.5)": kl_08_05, "KL(0.8||0.8)": kl_08_08}
```

## Exercise 6. Mutual information as KL divergence


Substituting `p(x,y)` and `p(x)p(y)` into the KL definition gives the usual mutual-information sum. Since KL is nonnegative, mutual information is also nonnegative and is zero exactly at independence.


## Exercise 7. Binary symmetric channel


```python
eps = sp.symbols("eps", positive=True)
binary_entropy = -eps * sp.log(eps) - (1 - eps) * sp.log(1 - eps)
I = sp.log(2) - binary_entropy
{"H(X)": sp.log(2), "H(X|Y)": binary_entropy, "I(X;Y)": I}
```

As `eps -> 0`, the channel becomes noiseless and `I(X;Y) -> log 2`. As `eps -> 1/2`, the output becomes independent noise and the mutual information goes to `0`.


## Exercise 8. Data processing inequality


If `X -> Z -> Y_hat` is a Markov chain, downstream processing through `h` cannot create information about `X` that `Z` no longer contains. Over-aggressive feature engineering can therefore erase predictive structure before the classifier sees it.


## Exercise 9. KL term in a VAE objective


Removing the KL term lets the encoder spread latent codes arbitrarily and usually destroys a smooth, sampleable latent space. If the KL term dominates too strongly, the posterior collapses toward the prior and the latent variables stop carrying useful information about `x`.


## Exercise 10. Differential entropy of a Gaussian


```python
sigma2, mu = sp.symbols("sigma2 mu", positive=True)
h = sp.Rational(1, 2) * sp.log(2 * sp.pi * sp.E * sigma2)
{"h(X)": h, "depends_on_mu": sp.diff(h, mu) if mu in h.free_symbols else 0}
```

The mean only shifts the density without changing its spread, so differential entropy depends on variance but not on location.
